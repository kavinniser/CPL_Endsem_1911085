```python
x = [1.5,1.8,2.1,2.7,3.0,3.3,3.6,3.9,4.8,5.1,5.4,5.7,6.0,6.6,6.9,7.2,7.8,8.1,8.4,8.7]
y = [0.45,0.42,0.42,0.36,0.33,0.31,0.32,0.31,0.28,0.29,0.24,0.23,0.25,0.21,0.23,0.21,0.23,0.22,0.24,0.23]
import math
#(i) y = (sigma1_x) e^(alpha1 x). Take ln both sides. lny = c1 + c2 x
lny = [math.log(i) for i in y]
linearfit(x,lny)


```

```python
# the slope = ln(sigma1_0) ; intercept = alpha1
sigma1_0 = math.exp(-0.0971676482997938)
alpha1 = -0.7720858223962891
print("The value of sigma_0 = " + str(sigma1_0))
print("The value of alpha = " + str(alpha1))
```

```python
t = np.linspace(0,10,3000,endpoint= True)
s = -0.7720858223962891 + t*-0.0971676482997938
plt.figure(figsize = (12, 8))
plt.plot(x,lny,'bo--')
plt.plot(t,s,'g', label='Precise Soln')
plt.xlabel(" $x$")
plt.ylabel("ln $(y)$")
plt.show()
```

```python
#(ii) y = (sigma2_x) T^(alpha2). Take ln both sides. lny = c3 + c4 lnx
lny = [math.log(i) for i in y]
lnx = [math.log(i) for i in x]
linearfit(lnx,lny)
```
